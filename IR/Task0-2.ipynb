# Concurrent

```python
import concurrent.futures
nums = [1, 1000001,2000001,3000001,4000001,5000001,6000001,7000001,8000001,9000001]
def sum(first):
    second = first + 999999
    return (first + second) * (second - first + 1)/2

with concurrent.futures.ThreadPoolExecutor(max_workers=10) as executor:
    futures = [executor.submit(sum, num) for num in nums]
    res = 0
    for future in concurrent.futures.as_completed(futures):
        res += future.result()

print(res)
```

```python
urls = ['https://api.github.com', 'https://api.github.com','https://api.github.com']
```

```python
def res(url):
    if not url:
        return 'N/A'
    r = requests.get(url)
    jsobj = r.json()

    if 'notifications_url' not in jsobj:
        return 'N/A'
    return jsobj['notifications_url']

with concurrent.futures.ThreadPoolExecutor(max_workers=3) as executor:
    futures = [executor.submit(res, url) for url in urls]
    rtn = []
    for future in concurrent.futures.as_completed(futures):
        rtn.append(future.result())

rtn
```

```python
with concurrent.futures.ProcessPoolExecutor(max_workers=10) as executor:
    futures = [executor.submit(sum, num) for num in nums]
    res = 0
    for future in concurrent.futures.as_completed(futures):
        res += future.result()

print(res)
```

```python
def res(url):
    if not url:
        return 'N/A'
    r = requests.get(url)
    jsobj = r.json()

    if 'notifications_url' not in jsobj:
        return 'N/A'
    return jsobj['notifications_url']

def main():
    with concurrent.futures.ProcessPoolExecutor(max_workers=3) as executor:
        futures = [executor.submit(res, url) for url in urls]
        rtn = []
        for future in concurrent.futures.as_completed(futures):
            rtn.append(future.result())
    
    print(rtn)

if __name__ == '__main__':
    main()
```
